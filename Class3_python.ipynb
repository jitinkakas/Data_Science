```python
data = np.array([
    [1, 85, 5, 90],
    [2, 78, 4, 85],
    [3, 92, 6, 95],
    [4, 60, 2, 70],
    [5, 75, 3, 80]
    ])
print(data)
print(data.shape)
print(data[0])
print(data[:,1]) # Print index 1 colomun
print(data[1:4]) # Print specific value metrix or rows
```

```python
data[:, 1] = data[:, 1] + 5
print(data)
```

```python
# numeric operations
print(np.mean(data[:,1]))
print(np.max(data[:,1]))
print(np.std(data[:,1]))
```

```python
high_scores = data[data[:,1] > 80] #Pipeline, Aggregation
print(high_scores)
```

```python
"""
Aggregation:-
    Aggregation refers to collecting multiple values and computing a single summary result from them.
    It's a fundamental concept used in data processing, databases, and OOP.
"""
```

```python
marks = data[:, 1]
print(marks.reshape(5, 1))
print(type(marks))
```

```python
import pandas as pd 
df = pd.DataFrame(data, columns = ["ID", "Marks", "Hours", "Attendance"])
print(df)
print(type(df))
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df["Performance"] = df["Marks"] > 80
print(df)
```

```python
sorted_df = df.sort_values(by = "Marks", ascending = False)
print(sorted_df)
```

```python
df.loc[2, "Marks"] = None
print(df)
```

```python
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
print(df)
```

```python
df["Category"] = ["A", "B", "A", "B", "A"]
print(df)
```

```python
print(df.groupby("Category")["Marks"].mean())
```
