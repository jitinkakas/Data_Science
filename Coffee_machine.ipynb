```python

from abc import ABC, abstractmethod

class Coffee(ABC):

    def __init__(self, name, price):
        # These are shared by ALL coffees
        self.name  = name
        self.price = price

    # @abstractmethod means every child class MUST implement this
    @abstractmethod
    def get_recipe(self):
        pass   # No body here — child classes will fill this in

    @abstractmethod
    def get_art(self):
        pass   # Each coffee will have its own ASCII art

    # This method is the same for all — so we define it here
    def display(self):
        print("\n" + "=" * 45)
        print(f"        ☕  {self.name.upper()}")
        print("=" * 45)
        print(self.get_art())           # Show the hash/ASCII image
        print("\n📋 RECIPE:")
        for step in self.get_recipe():  # Print each recipe step
            print(f"   {step}")
        print(f"\n💰 Price  : Rs.{self.price}")
        print("=" * 45)


# ── 1. LATTE ─────────────────────────────────
class Latte(Coffee):

    def __init__(self):
        # Call the parent class (Coffee) constructor
        super().__init__(name="Latte", price=180)

        # Private/encapsulated data — hidden inside the class
        self.__milk_ratio   = "3:1 (milk to espresso)"
        self.__temperature  = "65 C"
        self.__foam         = "Light microfoam"

    def get_recipe(self):
        return [
            "1. Pull a double espresso shot",
            f"2. Steam milk to {self.__temperature}  ->  {self.__milk_ratio}",
            "3. Pour steamed milk over espresso",
            f"4. Top with {self.__foam}",
            "5. Optional: Add latte art on top"
        ]

    def get_art(self):
        return """
        ( (       )  ) )
         ) )     (  ( (
        #################
        #               #
        #   ~ L A T T E #
        #    ~ ~ ~ ~ ~  #
        #  milk & love  #
        #_______________#
              | |
        """


# ── 2. MATCHA LATTE ──────────────────────────
class MatchaLatte(Coffee):

    def __init__(self):
        super().__init__(name="Matcha Latte", price=210)

        self.__matcha_grade = "Ceremonial grade"
        self.__origin       = "Uji, Japan"
        self.__milk_type    = "Oat milk"

    def get_recipe(self):
        return [
            f"1. Sift {self.__matcha_grade} matcha powder (from {self.__origin})",
            "2. Add 2 tbsp hot water (not boiling -- 80 C)",
            "3. Whisk matcha into a smooth paste",
            f"4. Steam {self.__milk_type} until frothy",
            "5. Pour milk over matcha paste",
            "6. Stir gently and enjoy the earthy sweetness!"
        ]

    def get_art(self):
        return """
        ( (        )  ) )
         ) )      (  ( (
        #################
        #  ~ ~ ~ ~ ~ ~  #
        #  M A T C H A  #
        #  ~ ~ ~ ~ ~ ~  #
        #   green gold  #
        #_______________#
              | |
        """


# ── 3. AMERICANO ─────────────────────────────
class Americano(Coffee):

    def __init__(self):
        super().__init__(name="Americano", price=120)

        self.__shots       = 2
        self.__water_ratio = "1:2 (espresso to water)"
        self.__strength    = "Bold & Strong"

    def get_recipe(self):
        return [
            f"1. Pull {self.__shots} shots of espresso",
            "2. Heat water to 90 C",
            f"3. Pour hot water over espresso  ->  Ratio: {self.__water_ratio}",
            f"4. Strength level: {self.__strength}",
            "5. Serve black -- no milk, no sugar (purist style!)"
        ]

    def get_art(self):
        return """
        ( (        )  ) )
         ) )      (  ( (
        #################
        #               #
        # A M E R I C  #
        #   - A N O -   #
        #  bold & pure  #
        #_______________#
              | |
        """


# ── 4. CAPPUCCINO ────────────────────────────
class Cappuccino(Coffee):

    def __init__(self):
        super().__init__(name="Cappuccino", price=160)

        self.__foam_thickness = "2 cm thick"
        self.__ratio          = "1:1:1 (espresso:milk:foam)"
        self.__serving_size   = "150 ml"

    def get_recipe(self):
        return [
            "1. Pull one espresso shot",
            f"2. Steam milk -- equal parts ratio: {self.__ratio}",
            "3. Pour steamed milk gently over espresso",
            f"4. Spoon thick foam on top -> {self.__foam_thickness}",
            "5. Dust with cocoa powder",
            f"6. Serve in {self.__serving_size} ceramic cup"
        ]

    def get_art(self):
        return """
        ( (        )  ) )
         ) )      (  ( (
        #################
        # ~~~ foam ~~~  #
        # C A P P U C  #
        #  - C I N O -  #
        #  rich & warm  #
        #_______________#
              | |
        """


# ── 5. ESPRESSO ──────────────────────────────
class Espresso(Coffee):

    def __init__(self):
        super().__init__(name="Espresso", price=100)

        self.__grind_size  = "Fine grind"
        self.__pressure    = "9 bars"
        self.__brew_time   = "25-30 seconds"
        self.__volume      = "30 ml shot"

    def get_recipe(self):
        return [
            f"1. Use {self.__grind_size} coffee beans",
            "2. Pack (tamp) grounds firmly into portafilter",
            f"3. Brew at {self.__pressure} pressure",
            f"4. Extraction time: {self.__brew_time}",
            f"5. Result: {self.__volume} of pure concentrated coffee",
            "6. Serve immediately -- espresso waits for no one!"
        ]

    def get_art(self):
        return """
          ( )   ( )   ( )
           ) ) ( ( ) ( (
        #################
        #               #
        # E S P R E S  #
        #   - S O -     #
        #  tiny & bold  #
        #_______________#
              | |
        """


# ─────────────────────────────────────────────
# STEP 3 — COFFEE MACHINE (Puts it all together)
# ─────────────────────────────────────────────

class CoffeeMachine:

    def __init__(self):
        # Store all coffees in a dictionary -- key = user choice number
        self.__menu = {
            "1": Latte(),
            "2": MatchaLatte(),
            "3": Americano(),
            "4": Cappuccino(),
            "5": Espresso(),
        }

    def show_menu(self):
        print("\n" + "#" * 45)
        print("#    Welcome to the Coffee Machine !!    #")
        print("#" * 45)
        print("   Select your coffee:\n")
        for key, coffee in self.__menu.items():
            print(f"   [{key}]  {coffee.name}  --  Rs.{coffee.price}")
        print("\n   [0]  Exit")
        print("#" * 45)

    def run(self):
        while True:
            self.show_menu()
            choice = input("\n   Enter your choice: ").strip()

            if choice == "0":
                print("\n   Thanks for visiting! Goodbye!\n")
                break
            elif choice in self.__menu:
                selected_coffee = self.__menu[choice]
                selected_coffee.display()    # Polymorphism in action!
                input("\n   Press Enter to go back to menu...")
            else:
                print("\n   Invalid choice! Please try again.")


# ─────────────────────────────────────────────
# STEP 4 — RUN THE PROGRAM
# ─────────────────────────────────────────────

if __name__ == "__main__":
    machine = CoffeeMachine()
    machine.run()
```
